```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('WorldCupMatches.csv')
df.head(10)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
# @title Frequency of Away Team Goals

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
sns.histplot(df['Away Team Goals'], bins=20, kde=True, ax=ax)

sns.despine(fig=fig, ax=ax)
plt.title('Distribution of Away Team Goals')
plt.xlabel('Away Team Goals')
plt.ylabel('Frequency')
plt.show()
```

```python
# @title Year vs Home Team Goals

df_sorted = df.sort_values('Year', ascending=True)

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
sns.lineplot(data=df_sorted, x='Year', y='Home Team Goals', marker='o', ax=ax)
sns.despine(fig=fig, ax=ax)

plt.xlabel('Year')
plt.ylabel('Home Team Goals')
plt.title('Year vs Home Team Goals')
plt.show()
```

```python
# @title Attendance

from matplotlib import pyplot as plt
df['Attendance'].plot(kind='line', figsize=(8, 4), title='Attendance')
plt.gca().spines[['top', 'right']].set_visible(False)
```
